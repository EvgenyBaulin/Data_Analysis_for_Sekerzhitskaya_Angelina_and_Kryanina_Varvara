```python
import numpy as np
import matplotlib.pyplot as plt

# Генерация случайных данных с корреляцией -0.05
np.random.seed(42)
x = np.random.rand(100)
y = -1 * x + np.random.normal(scale = 0.1, size = 100)

# Вычисление коэффициента корреляции
correlation_coefficient = np.corrcoef(x, y)[0, 1]

# Построение графика
plt.scatter(x, y, label = f'Корреляция: {correlation_coefficient:.2f}', alpha = 0.8)
plt.title('График с коэффициентом корреляции -0.05')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.grid(True)
plt.show()
```
